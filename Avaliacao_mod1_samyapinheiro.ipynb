```python
# importar pacotes #
import os
from pathlib import Path

### Biblioteca ###
class biblioteca():
    
    def __init__(self, nome):
        
        #organizando nomes diretorio
        raiz = Path(os.getcwd())
        nome = nome.lower()
        nome = nome.replace(' ', '')
        self.path = Path(raiz) / Path(nome)
        
        #verificar existência da biblioteca
        try:
            os.mkdir(self.path)
        except FileExistsError:
            pass
        
        #criar diretórios temas
        self.esporte = self.__criar_pasta_tema__('esporte')
        self.politica = self.__criar_pasta_tema__('politica')
        self.tecnologia = self.__criar_pasta_tema__('tecnologia')
        
        
        
    def __criar_pasta_tema__(self, tema):
        tema_lower = tema.lower() #padronizar diretorios
        caminho_pasta = self.path / Path(tema_lower)
        #verificar existência diretorio
        try:
            os.mkdir(caminho_pasta)
        except FileExistsError:
            pass
        
        return caminho_pasta
  
```

```python
# importar pacotes #
import os
from pathlib import Path
from datetime import datetime

### Indice ###
class indice():
    
    def __init__(self):
        self.path = self.__criar_indice_desorganizado__()
        self.total_artigos = self.__checar_qtos_artigos__()

    def __criar_indice_desorganizado__(self):   
        #cria arquivo de índice (sobreescreve se existir)
        path_indice = Path(os.getcwd()) / Path('indice_biblioteca.txt')
        with open(path_indice, 'w', encoding = 'UTF-8') as arq_indice:
            header = 'Data (dd/mm/AAAA) --   Assunto  -- Título\n'
            arq_indice.write(header)
        return path_indice
    
    def __checar_qtos_artigos__(self):
        #ler indice
        arquivo_indice = open(self.path, 'r', encoding='utf-8')
        conteudo = arquivo_indice.readlines()
        arquivo_indice.close()
        #contar total de artigos
        total_artigos = len(conteudo) - 1
        return total_artigos

    def cadastrar_artigo(self, data_artigo, tema_artigo,  titulo_artigo):
        #cria linha formatada
        linha = f'{data_artigo:17} -- {tema_artigo:10} -- {titulo_artigo}\n'
        #append linha formatada em self.path
        with open(self.path, 'a', encoding = 'UTF-8') as arq_indice:
            arq_indice.write(linha)
    
    def organizar_indice(self):
        #ler indice
        arquivo_indice = open(self.path, 'r', encoding='utf-8')
        conteudo = arquivo_indice.readlines()
        arquivo_indice.close()
        
        #organizar lista por artigo
        lista_artigos = []
        for i in range(len(conteudo)):
            lista_artigos.append(conteudo[i].replace('\n', '').replace('--', '').split())
        return lista_artigos[1:]
    
    def ordenar_indice(self):
        lista_artigos = self.organizar_indice()
        lista_datas = [lista_artigos[i][0].replace(' ','') for i in range(0,len(lista_artigos))]
        lista_datas = [datetime.strptime(lista_datas[i], '%d/%m/%Y') for i in range(0,len(lista_datas))]
        
        #ordenar datas e retorna ordenação para indice
        lista_datas_ordem = []

        for i, date in enumerate(lista_datas):

            if i == 0:
                lista_datas_ordem.append([i, date])
            else:

                eh_maior = date >= lista_datas_ordem[-1][1]

                if eh_maior:
                    lista_datas_ordem.append([i, date])
                else:
                    lista_datas_ordem.insert(0,[i, date])

        ordem_lista_ori = [lista_datas_ordem[i][0] for i in range(len(lista_datas_ordem))]
        return ordem_lista_ori
             
    def imprimir_indice_ordenado(self):
        ordem_lista_ori = self.ordenar_indice()
        
        #abrir e imprimir ordenado
        arquivo_indice = open(self.path, 'r', encoding='utf-8')
        conteudo = arquivo_indice.readlines()
        arquivo_indice.close()
        
        print(conteudo[0])
        for i in ordem_lista_ori:
            j = i + 1 #adiciona um para pular header
            print(conteudo[j]) 
```

```python
# importar pacotes #
import os
from pathlib import Path

### Artigo ###
class artigo():
    def __init__(self, data_artigo, titulo_artigo, tema_artigo, conteudo_cru, biblioteca):
        self.data = data_artigo
        self.titulo = titulo_artigo
        self.tema = tema_artigo
        self.texto = conteudo_cru
        self.__salvar_artigo__(biblioteca)
        
    def __salvar_artigo__(self, biblioteca):
        nome = self.tema.lower() + '_' + self.titulo[:5] + self.data.replace('/', '') + '.txt'
        self.path = biblioteca.path / self.tema.lower() / Path(nome)
        print(self.path)
        
        #abre arquivo e excreve o conteúdo
        with open(self.path, 'w', encoding = 'UTF-8') as arq_artigo:
            header = f'{self.data:17} -- {self.tema:10} -- {self.titulo}\n'
            arq_artigo.write(header)
            arq_artigo.write(self.texto)
    
    def __texto_minuscula__(self):
        return self.texto.lower()
    
    def __remove_pontuacao__(self, texto):       
        pontuacoes = [',', '.', '?', '!', '(', ')', '[', ']', ';', ':']
        
        for pontuacao in pontuacoes:
            texto = texto.replace(pontuacao, '')
        
        return texto
    
    def __tokeniza__(self, texto):
        return texto.split()
    
    def __remove_stopword__(self, lista_tokens):       
        #importar arquivo de stop word
        arquivo_stop = open('stopword.txt', 'r', encoding='utf-8')
        conteudo = arquivo_stop.readlines()
        arquivo_indice.close()
        #limpar stopwords
        stopword = [conteudo[i].replace('\n', '') for i in range(len(conteudo))]
        stopword = [stopword[i].replace(' ', '') for i in range(len(conteudo))]
        
        lista_limpa = [palavra for palavra in lista_tokens if palavra not in stopword]
        
        return lista_limpa
    
    def __conta_palavras__(self, lista_tokenslimpa):
        lista_palavras = []
        lista_contagem_palavras = []

        for token in lista_tokenslimpa:
            if token not in lista_palavras:
                lista_palavras.append(token)
                lista_contagem_palavras.append(lista_tokenslimpa.count(token))
        #formata como dicionário        
        return dict(zip(lista_palavras, lista_contagem_palavras))
        
    def dic_palavras_artigo(self):
        
        texto = self.__texto_minuscula__()
        texto = self.__remove_pontuacao__(texto)
        lista_tokens = self.__tokeniza__(texto)
        lista_limpa = self.__remove_stopword__(lista_tokens)
        
        self.palavras = self.__conta_palavras__(lista_limpa) 
        
    
```
