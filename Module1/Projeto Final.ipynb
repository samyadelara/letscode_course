```python
# API para renderizar grficos em java
# API sempre veja a documentao
# QuickChart.io

def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return [
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false' #api pede assim e nao False
    return {
        'title': title,
        'display': display
    }
    
```

```python
def create_chart(x, y, labels, kind='bar', title=''):
    
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
# imagem em content  binrio
```

```python
def save_image(path, content):
    with open(path, 'wb') as image: #b  binrio
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []
for obs in final_data[1::10]:
    y_data_1.append(obs[CONF])
    
y_data_2 = []
for obs in final_data[1::10]:
    y_data_2.append(obs[RECUP])
    
labels = ['Confirmados', 'Recuperados']
    
x = []
for obs in final_data[1::10]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_1, y_data_2],
                    labels, 
                    title='Grfico confirmados vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('meu-primeiro-grafico.png')
```
