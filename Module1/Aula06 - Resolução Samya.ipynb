```python
# tokenização #
# 1 - retirar pontuação
pontuacao = [',', '.', ':', ';', '!', '\n', '?', '(', ')']

i = pontuacao[0]
texto1 = texto_fonte.replace(i, '')

for i in pontuacao[1:]:
    texto1 = texto1.replace(i, '')

#print(texto1)

# 2 - limpeza do texto
texto2 = texto1.lower()


# 3 - tokenização
lista_palavras = texto2.split()

#print(lista_palavras)
```

```python
# 4 - Achar ocorrências únicas
lista_palavras_unicas = []

for palavra in lista_palavras:
    palavra_nova = palavra not in lista_palavras_unicas
    if palavra_nova:
        lista_palavras_unicas.append(palavra)

#print(lista_palavras_unicas)
```

```python
# 5 - Contar ocorrências de palavras únicas na lista de palavras original

ocorrencias = []

for palavra in lista_palavras_unicas:
    cont = lista_palavras.count(palavra)
    ocorrencias.append(cont)

#print(ocorrencias)
```

```python
# 6 - Lista única com Dados palavras_unicas + contagem

lista_resultado = [lista_palavras_unicas, ocorrencias]
#lista_resultado
```

```python
# 7 - Criando a tabela html
inicio = '\n<ul class="cloud">\n'
recheio_html = ''

for i in range(len(ocorrencias)):
    linha = f'  <li><a data-weight="{lista_resultado[1][i]}">{lista_resultado[0][i]}</a></li>\n'
    recheio_html = recheio_html + linha

fim = '</ul>'
tabela_html = inicio + recheio_html + fim

print(tabela_html)

```
